```python
import pandas as pd
```

```python
def make_var_info(cite, tem):

    df_cite = pd.read_csv(cite)
    df_cite = df_cite.set_index('SynPE ref\nInput')

    df_tem = pd.read_csv(tem)
    df_tem = df_tem.set_index('RefSeq')

    list_aa_var  = []
    list_snv_var = []

    for ref in df_tem.index:
        try:
            data = df_cite.loc[ref]

            list_aa_var.append(data.Control)
            list_snv_var.append(data.Variant)

        except:
            list_aa_var.append('-')
            list_snv_var.append('-')

    df_tem['AA_var'] = list_aa_var
    df_tem['SNV_var'] = list_snv_var

    return df_tem
```

```python
cite = 'Exon9_KXC_Y9_B_output_231004.csv'
tem  = 'variants_info/ex9_output_template.csv'

df_info = make_var_info(cite, tem)

df_info.to_csv('variants_info/ex9_info.csv')
```

```python
def make_var_info_for_ex4(cite, tem):

    df_cite = pd.read_csv(cite)
    df_cite = df_cite.set_index('SynPE_ref_R2').drop_duplicates()

    df_tem = pd.read_csv(tem)
    df_tem = df_tem.set_index('RefSeq')

    list_aa_var  = []
    list_snv_var = []

    for ref in df_tem.index:

        try:
            
            data = df_cite.loc[ref]

            if type(data) == pd.Series:
                list_aa_var.append(data.Control)
                list_snv_var.append(data.Variant)

            else:
                data = data.iloc[0]
                list_aa_var.append(data['Control'])
                list_snv_var.append(data['Variant'])

        except:
            list_aa_var.append('-')
            list_snv_var.append('-')

    df_tem['AA_var'] = list_aa_var
    df_tem['SNV_var'] = list_snv_var

    return df_tem
```

```python
cite = 'unedit_ref.csv'
tem  = 'variants_info/ex4-2_output_template.csv'

df_info = make_var_info_for_ex4(cite, tem)

df_info.to_csv('variants_info/invivo_ex4_info.csv')
df_info
```
